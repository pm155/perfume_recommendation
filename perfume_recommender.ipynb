```python
def extract(obj):
    l=[]
    obj = obj.split(",")
    for i in obj:
        l.append(i.replace(" ",""))
    return l
    
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x)) #converting to strings
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags'])
```

```python
vector = vector.toarray()
```

```python
vector
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(vector)
```

```python
sim.shape
```

```python
def recommender(per):
    index = df[df['Fragrance'] == per].index[0]
    distances = sorted(list(enumerate(sim[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].Fragrance)
        
```

```python
recommender('Tihota Eau de Parfum - Indult')
```
